## FOI OBSERVADO QUE O CAMPO "data" ESTÁ EM str(string) - TRANSFORMAR EM FORMATO DE DATA

```python
df_movi_func["data"] = pd.to_datetime(df_movi_func["data"], format="%d/%m/%Y")

df_movi_func.info()
```

## REALIZAR ALGUNS FILTROS NA PLANILHA

```python
ft_cargo = df_movi_func[df_movi_func["descricaoFuncionarioSindicato"] == "CAESB (GDOC)"]

ft_escala = df_movi_func[df_movi_func["descricaoFuncionarioEscalaTrabalho"] == "SEG,TER,QUA,QUI,SEX - 07:00:00-13:00:00"]

ft_cargo_escala = df_movi_func[
    (df_movi_func["descricaoFuncionarioSindicato"].isin(["ANVISA REC", "CAESB (GDOC)"])) 
    & (df_movi_func["descricaoFuncionarioEscalaTrabalho"] == "SEG,TER,QUA,QUI,SEX - 07:00:00-13:00:00")
    ]

#ft_cargo.head(50)
#ft_escala.head(100)
ft_cargo_escala.head(100)
```
